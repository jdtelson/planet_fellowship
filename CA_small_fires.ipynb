```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fires = pd.read_csv('mapdataall.csv')
fires
```

```python
placer = fires[fires['incident_county']=='Placer'][fires['incident_date_created'].str.match('202.*')]
```

```python
nevada = fires[fires['incident_county']=='Nevada'][fires['incident_date_created'].str.match('202.*')]
```

```python
el_dorado = fires[fires['incident_county']=='El Dorado'][fires['incident_date_created'].str.match('202.*')]
```

```python
three_counties = placer.append(nevada).append(el_dorado)
```

```python
three_counties.to_csv('placer_nevada_elDorado_fires.csv')
```
